```python
import pandas as pd
import numpy as np

titanic = pd.read_csv("/home/deven/Music/titanic_train.csv")
```

```python
titanic.head()
```

```python
titanic.drop('Name',axis =1,inplace =True)
```

```python
df = titanic.copy()
```

```python
df
```

```python
def temp(x):
    if(x == "Male"):
        return 1
    else :
        return 0
    
df["sex"]=df.Sex.apply(temp)
```

```python
del df['Sex']
```

```python
df.Cabin.fillna('CS85',inplace =True)

df
```

```python
df.Age.fillna(df.Age.mean(),inplace =True)
print(df)
```

```python
df
```

```python
def fun(t):
    if(t == 'Q'):
        return 1
    elif(t== 'S'):
        return 0
    elif(t =='C'):
        return -1

df["Embark"] = df.Embarked.apply(fun)
```

```python
df.drop('Embarked',axis =1,inplace =True)
```
